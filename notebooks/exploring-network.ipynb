```python
def add_bus_node(node_code, row, bus_dict):
    '''
    Basic function which adds a bus node if there is not a matching one if there is not one in
    bus_dict already.
    '''
    for idx,n in enumerate(bus_dict["name"]):
        if node_code[3:6] == n[3:6]: # 3-alpha match on first match.. nothing too complicated yet..
            bus_dict["name"].append(node_code) # add name based on the line node it is attached to.
            bus_dict["v_nom"].append(row["voltage_in_kv"]) # default to be replaced with line data later....
            bus_dict["x"].append(bus_dict["x"][idx])
            bus_dict["y"].append(bus_dict["y"][idx])
            bus_dict["carrier"].append("AC")
            bus_dict["node_type"].append(bus_dict["node_type"][idx])
            break
        else:
            continue

    # bus_dict["name"].append(row["node_code"])
    # bus_dict["v_nom"].append(230) # default to be replaced with line data later....
    # bus_dict["x"].append(row["longitude"])
    # bus_dict["y"].append(row["latitude"])
    # bus_dict["carrier"].append("AC")
```

```python
# deduplicate
# cols_dedup = list(set(df_lines_bc.columns.tolist()) - set(['transmission_line_id']))
# df_lines_bc.drop_duplicates(subset=cols_dedup, keep=False, inplace=True)


# setup
bus_dict = {
        'name':[],
        'v_nom':[],
        'x':[],
        'y':[],
        'carrier':[],
        "node_type":[],
    }   

line_dict = {
        'name': [],
        'type': [],
        'bus0': [],
        'bus1': [],
        'length': [],
        's_nom': [],
        'v_coders':[],
    }
# (1) iterate over buses
for idx,row in df_sub_bc.iterrows():
    bus_dict["name"].append(row["node_code"])
    bus_dict["v_nom"].append(230) # default to be replaced with line data later....
    bus_dict["x"].append(row["longitude"])
    bus_dict["y"].append(row["latitude"])
    bus_dict["carrier"].append("AC")
    bus_dict["node_type"].append(row["node_type"])

 
# (2) iterate over lines then apply node names based on buses
unique_3_alpha_buses = [n[3:6] for n in bus_dict["name"]]
# Methods for matching lines start/end nodes to nodes within the substation dataset
# 1) Identical match
# 2) Find a substation node with  the middle 3 characters matching (i.e. AAL in BC_AAL_DSS)
#    Use this substations' location to create a new bus named after the node_coe

tracker = [] # list to track buses added to dataset
for idx,row in df_lines_bc.iterrows():
    # Check to make sure this line has a potential match 3-alpha match (i.e. This will take out BC_WAX_GSS as an example)
    if (row["starting_node_code"][3:6] in unique_3_alpha_buses) and (row["ending_node_code"][3:6] in unique_3_alpha_buses): # ensure there is a 3-alpha match for both end nodes
        line_dict["name"].append(row["starting_node_code"][3:] + row["ending_node_code"][2:]) # XXX_GSS_CCC_IPT
        line_dict["type"].append(f'{str(int(row["voltage_in_kv"]))}kV') # Add voltage of the line
        
        if row["starting_node_code"] in bus_dict["name"]:
            line_dict["bus0"].append(row["starting_node_code"])
            # fix voltage of bus_dict["v_nom"] 
        else:
            add_bus_node(row["starting_node_code"], row, bus_dict)
            line_dict["bus0"].append(row["starting_node_code"])

        if row["ending_node_code"] in bus_dict["name"]:
            line_dict["bus1"].append(row["ending_node_code"])
        else:
            add_bus_node(row["ending_node_code"], row, bus_dict)
            line_dict["bus1"].append(row["ending_node_code"])

        line_dict["length"].append(row["line_segment_length_km"]) # Add voltage of the line
        line_dict["s_nom"].append(3)
        line_dict["v_coders"].append(row["voltage_in_kv"])

    

# # Save buses
# pd.DataFrame(bus_dict).to_csv("pypsa-network/buses.csv",index=False)

# # Save lines
# # need to look into deuplicates in more detail
# pd.DataFrame(line_dict).drop_duplicates(subset=["name"]).to_csv("pypsa-network/lines.csv",index=False)
```
